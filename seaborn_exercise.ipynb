### Task 1: Plot a regression line between attack and defense.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv', sep=',')
df.drop('#', axis=1, inplace=True)

# Plot a regression line between attack and defense
plt.figure(figsize=(10, 6))
sns.regplot(x='Attack', y='Defense', data=df)
plt.title('Regression Line between Attack and Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()
```

### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv', sep=',')
df.drop('#', axis=1, inplace=True)

# Plot a regression line between attack and defense with different generations
plt.figure(figsize=(12, 8))
sns.lmplot(x='Attack', y='Defense', hue='Generation', data=df, aspect=1.6, markers='o', scatter_kws={'s': 50, 'alpha': 0.5})
plt.title('Regression Line between Attack and Defense by Generation')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.grid(True)
plt.show()
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv', sep=',')
df.drop('#', axis=1, inplace=True)

# Drop the 'Generation' and 'Legendary' columns
df = df.drop(columns=['Generation', 'Legendary'])

# Melt the DataFrame to long format
df_melted = df.melt(var_name='Feature', value_name='Value')

# Plot the boxplots
plt.figure(figsize=(15, 10))
sns.boxplot(x='Feature', y='Value', data=df_melted)
plt.title('Boxplot of Each Feature')
plt.xlabel('Feature')
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.show()
```

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv', sep=',')
df.drop('#', axis=1, inplace=True)

# Drop the 'Generation' and 'Legendary' columns
df = df.drop(columns=['Generation', 'Legendary'])

# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Pokémon DataFrame')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv', sep=',')
df.drop('#', axis=1, inplace=True)

# Drop the 'Generation' and 'Legendary' columns
df = df.drop(columns=['Generation', 'Legendary'])

# Check the data types of the DataFrame
print(df.dtypes)

# Select only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = df_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Pokémon DataFrame')
plt.show()
```
