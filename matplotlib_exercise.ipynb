### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from io import StringIO

# URL of the dataset
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

# Fetch the data
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
data = response.text

# Load the data into a pandas DataFrame
df = pd.read_csv(StringIO(data))

# Extract the 'total_profit' column
total_profit = df['total_profit']

# Plot the 'total_profit' using a line plot
plt.figure(figsize=(10, 5))
plt.plot(total_profit, marker='o')
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from io import StringIO

# URL of the dataset
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

# Fetch the data
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
data = response.text

# Load the data into a pandas DataFrame
df = pd.read_csv(StringIO(data))

# List of product sales columns
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Plot the sales data for all products
plt.figure(figsize=(12, 6))
for column in product_columns:
    plt.plot(df['month_number'], df[column], marker='o', label=column)

plt.title('Sales Data of Different Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from io import StringIO

# URL of the dataset
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

# Fetch the data
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
data = response.text

# Load the data into a pandas DataFrame
df = pd.read_csv(StringIO(data))

# Extract the 'total_profit' column
total_profit = df['total_profit']

# Plot the 'total_profit' using a histogram
plt.figure(figsize=(10, 5))
plt.hist(total_profit, bins=10, edgecolor='black')
plt.title('Distribution of Total Profit')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
